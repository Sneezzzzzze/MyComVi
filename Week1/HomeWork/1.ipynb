```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# โหลดข้อมูล
df = pd.read_csv('https://raw.githubusercontent.com/lazyprogrammer/machine_learning_examples/master/tf2.0/moore.csv', header=None)

# เปลี่ยนชื่อคอลัมน์
df = df.rename(columns={0: "years", 1: "data"})

# เตรียมข้อมูลสำหรับการถดถอย
X = df['years'].values.reshape(-1, 1)  # แกน x
y = df['data'].values  # แกน y

# สร้างโมเดลการถดถอยเชิงเส้น
linear_model = LinearRegression()
linear_model.fit(X, y)

# สร้างกราฟกระจาย
plt.scatter(X, y, color='blue', label='Data Points')

# สร้างเส้นสมการจากโมเดล
y_pred = linear_model.predict(X)
plt.plot(X, y_pred, color='red', label='Linear Fit')

plt.title('Scatter Plot of Years vs Data with Linear Fit')
plt.xlabel('Years')
plt.ylabel('Data')
plt.legend()
plt.show()

# แสดงสมการ
intercept = linear_model.intercept_
slope = linear_model.coef_[0]
print(f"สมการ: y = {intercept} + {slope} * x")
```
